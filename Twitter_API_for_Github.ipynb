### CLIENT/TWEEPY

```python
import tweepy
```

```python
client = tweepy.Client(bearer_token=BEARER_TOKEN, wait_on_rate_limit=True)
```

1. Schritt: Vom Handle zur User-ID

```python
#id = "3069610055"
handle = "firefoxx66"
user = client.get_user(username=handle)
print(f"The user id for handle @{handle} is {user.data.id}.")
```

2. Die User-ID in den folgenden Code eingeben, um die Followers zu scrapen

```python


import time
import csv
a = []
#list_follower_laeri = 
paginator = tweepy.Paginator(
    client.get_users_followers,               # The method you want to use
    "23242375",                            # Some argument for this method
    #end_time="2022-03-31T11:59:59Z",       # Some argument for this method
    #exclude=['retweets', 'replies'],       # Some argument for this method
    #start_time="2021-03-31T11:59:59Z",     # Some argument for this method
    max_results=1000                        # How many tweets asked per request
)

try: 
    
    for i, follower in enumerate(paginator.flatten()):      # Default to inf
        if i%500 == 0:
            print(i)
            print(len(a))
        a.append(follower)
except tweepy.RateLimitError as exc:
    timestr = time.strftime("%Y%m%d-%H%M%S")
    outfile_name = "out_"+timestr+".csv"
    print('Rate limit! will save names at: {}, len b: {}'.format(
        outfile_name, len(a)))

    with open(outfile_name, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerows(a)
```

```python
df = pd.DataFrame(a)
```

3. Alles als Dataframe speichern, mit dem passenden Namen

```python
df.to_csv('hodcroft_list_of_followers.csv') 
```
